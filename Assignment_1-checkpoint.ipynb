```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
dataset = pd.read_csv(r"C:\Users\Asad shaikh\Documents\sales_data_with_discounts.csv")
```

```python
numerical_cols = dataset.select_dtypes(include=['float64', 'int64']).columns
numerical_cols
```

```python
dataset.describe()
```

```python
dataset[numerical_cols].hist(bins=20, figsize=(10, 8))
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=dataset[numerical_cols])
plt.title('Boxplots of Numerical Columns')
plt.show()
```

```python
categorical_cols = dataset.select_dtypes(include=['object']).columns

for col in categorical_cols:
    plt.figure(figsize=(8, 6))
    dataset[col].value_counts().plot(kind='bar')
    plt.title(f'Bar Chart of {col}')
    plt.ylabel('Count')
    plt.show()
```

```python
scaler = StandardScaler()
scaled_data = scaler.fit_transform(dataset[numerical_cols])
scaled_df = pd.DataFrame(scaled_data, columns=numerical_cols)

# Show before and after comparison
comparison = pd.concat([dataset[numerical_cols].head(), scaled_df.head()], axis=1)
comparison.columns = [f'Original_{col}' for col in numerical_cols] + [f'Scaled_{col}' for col in numerical_cols]
print(comparison)
```
